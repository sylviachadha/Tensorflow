# Tensorboard for visualization

Tensorboard is used to visualize the computation graph. This tool is not for DATA visualization, it is used by developers to visualize compuational graphs.

Tf visualization may be useful for complex networks like neural networks where you have hundreds of layers & 100's of neurons in each layer etc

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

YOu create each of these constants, variables, operations and give them a name, this name will be used by tensorboard when creating a graph.

**Create computational graph/tensorflow program**

```python
# Nodes in computational graph (constants, variables, ops all r nodes)
# # Define your TensorFlow graph

a = tf.constant(10, name='a')
b = tf.constant(20, name='b')
c = tf.add(a,b, name='c')
d = tf.multiply(a,b, name='d')
e = tf.multiply(c,d, name='e')
```

**View computational graph in tensorboard**

```python
sess = tf.Session()
```

```python
sess.run(e)
```

```python
writer = tf.summary.FileWriter('graph',sess.graph)
writer.close()
```

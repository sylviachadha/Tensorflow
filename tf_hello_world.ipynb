# Hello world using tensorflow 1.0 ( computational graph/code can only be executed within a session)

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
hello = tf.constant("Hello World")
```

```python
print(hello)    # it does not display o/p because u r just creating a computational graph here

                # u cannot execute it outside of a session
    
                # in this case, r graph has just 1 node "Hello"
```

```python
sess = tf.Session()
```

```python
sess.run(hello)
```

```python
sess.close()
```

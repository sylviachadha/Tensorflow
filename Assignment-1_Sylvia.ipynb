### 1. Find C = A X B

```python
# C = A @ B
C = tf.matmul(A, B)
print(C)
```

- C=A x B is possible as the shapes of tensors A and B are compatible to perform multiplication.
- A has 3 rows and 5 columns, shape= (3,5)
- B has 5 rows and 4 columns, shape= (5,4)
- Since number of columns of A which is 5 is equal to number of rows of B which is 5, hence multiplication is possible.


### 2. Is it possible to perform B X A. If not, give reason?

- It is not possible to perform B x A due to shape incompatibility.
- B is (5,4)
- A is (3,5)
- Since number of columns of B which is 4 is not equal to number of rows of A which is 3, hence incompatible to perform multiplication.

### **Implementing a function with pre-determined values**

Solving function like below using tensorflow where values of x,y,z are pre-determined like x=3, y=4 and c=2

We could have created X and Y as constants since values r known but just in case we want values of X and Y to be changed hence we define as Variable. We initialize these variables with X and Y with values of 3 and 5.

```python
X = tf.Variable(3)
```

```python
Y = tf.Variable(4)
```

C is a typical constant, we declare it as a constant.

```python
C = tf.constant(2)
```

Create expression for the function

```python
Z = X*X*Y+Y+C
```

```python
print(Z)
```

### **Implementing a function using user-input**

When instead of just 2 variables, you might need to read a million images, process becomes complicated,so u read maybe 100 images at a time.

In TensorFlow 2.0, placeholders have been deprecated and replaced with the tf.keras.Input layer. You can specify the batch size in tf.keras.Input by setting the first dimension of the shape argument to a fixed integer value or to None to allow for variable batch sizes. Shape argument is set to (None, 10), which means that the input tensor can have any batch size (i.e., it can vary from batch to batch) and 10 features.

**Passing multiple values to a function (define and call a function)**

```python
# define function using operator symbols
def calc1(X,Y):
    Z = X*X*Y+Y+C
    return Z
```

```python
# define function using tf
# Since every operation (ops) is one node hence u need to define multiply, add etc again
# and again as per the number of times the operation is performed.

def calc2(X,Y):
    Z1 = tf.multiply(tf.multiply(X,X),Y)
    print(Z1)
    Z2 = tf.add(tf.add(Z1,Y),C)
    return Z2
```

```python
# Call function
rzlt1 = calc1(3,4)
rzlt1
```

```python
# Call function
rzlt2 = calc2(3,4)
rzlt2
```

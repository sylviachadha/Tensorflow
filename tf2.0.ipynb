# Tensorflow 2.0

## tf1.0 vs tf2.0 changes


- Tensor1.0 is **Lazy execution** First you create/define a computation graph and then you execute comptational graph in a session.
- Tensor2.0 is **Eager execution** which means can write code like a normal python program & it gets executed right away, so there r no more sessions. Also variable initialization is not required.


tf1.0

Purpose of having and executing graph in sessions was to make code scalable so it can be executed in a distributed mode.

tf2.0

Use **tf.function also known as decorator** in tf2.0 for distributed mode since there are no more sessions in tf2.0. The Functionality u want to be executed in a distributed mode, u write it as a function and you annotate tf.function decorator to that. If decorator is added to taht function, tf understands that functionality needs to be executed in a distributed mode using a graph.

tf2.0 - Consolidation of high level API's. Lot of high level functions have been consolidated under a version of Keras API called tf.keras.

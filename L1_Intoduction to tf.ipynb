# Tensorflow library

### Tensors (matrices of higher dimensions) are part of tensorflow library.

1. Numpy library mainly for matrices. Matrix operations possible with numpy library. Some features of Numpy also seen in Pandas(for data analysis).

2. Tensorflow library particularly for image processing and it is particularly made up for different types of neural networks.

3. At core of everything is data. Data can be represented in form of numbers. You can have single entity like 5 or collection of entities like (3,4,5) or can have several samples like we had in form of tabular data (which is a collection of samples and attributes). So tabular data itself is made up of number of collections like S1=(3,4,5), S2=(7,8,9) etc. When u put all these S1, S2 together u get data in form of tabular format.This is known as 2 dimensional data.

4. We have seen data till 2D format but we can go to higher dimension. We can have data representation in 3D, 4D and even higher order formats, the complexity will increase.

5. Tensors - Whenever we say matrices which are of higher order, we say they r tensors. Tensorflow is an umbrella term for all types of entities if we talk in terms of rank, axis etc however more specifically they are used for matrices of higher order (referring to more than order 2).

6. Tensor operations r similar to matrices operations & tensor operations they form the core of image processing (higer order matrices).

7. In python when we say numeric value it can either be integer or float. So both integer & float r ok with tensors, they will be represented as numbers only & strings ofcourse r different.

8. Tensor Characteristic

- **Rank** (number of dimensions)
- **Shape** (a list that consists of length of each of the dimension)
- **Size** (total elements within a tensor)
- **Homogenous elements in a tensor**

9. Tensor is an umbrella term for different types of representation of data.



- Scalar is a fixed value/quantity, neither it does not have a rank or dimension.
- When we have a group of no's we used to have data structures where we can put several values together, those were vectors & they have rank 1 & dimension 1.
- Matrices will have 2 axes & it will have rank & dimension both as 2.

10. Examples of higher order data representations

- **Time Series data/ Stock Market data / Sequence Data** can be represented as a 3D data.
- **Image Data** - Images in themselves can be represented as a 3D vector (height, width, color depth). Image data is a 4D data (samples, height, width, channels).
- **Video Data** - It is 5D data where in addition to 4D image data we have to consider the frames as well.

11. How tf is better than numpy?

Although numpy can also be used for higher order matrices but tf is particularly prepared for this purpose. Tf is computationally faster. Memory management of tf is better than that of numpy.tf is built over numpy, so it is using features of numpy but in a better manner.

```python
import tensorflow as tf
```

```python
C1 = tf.constant(20)
C2 = tf.constant(30)
C3 = tf.constant(10)
```

**Using tensorflow functions**

```python
def equation1(X,Y):
    z1 = tf.pow(X, 3)                                          # x^3
    z2 = tf.pow(Y, 3)                                          # y^3
    z3 = tf.multiply(tf.multiply(tf.pow(X, 2),Y),C1)           # 20.x^2.y
    z4 = tf.multiply(tf.multiply(tf.pow(Y, 2),X),C2)           # 30.y^2.x
    z5 = tf.add(tf.add(tf.add(tf.add(z1,z2),z3),z4),C3)        # final Equation
    return z5
```

```python
rzlt1 = equation1(10,20)
rzlt2 = equation1(23,34)
rzlt3 = equation1(252,423)
print("Results are", rzlt1,rzlt2,rzlt3 )
```

**Using tensorflow operators**

```python
def equation2(X,Y):
    result = X*X*X + Y*Y*Y + C1*X*X*Y + C2*Y*Y*X + C3
    return result
```

```python
rzt1 = equation2(10,20)
rzt2 = equation2(23,34)
rzt3 = equation2(252,423)
print("Results are", rzt1,rzt2,rzt3 )
```

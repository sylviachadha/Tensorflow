# Constants, Variables and Placeholders

```python
import tensorflow as tf
```

**Constants**

```python
one = tf.constant(1)
```

```python
print(one)
```

**Variables**

```python
ten = tf.Variable(10)
```

```python
print(ten)
```

```python
sum1 = tf.add(ten,one)
sum1
```

```python
sum2 = ten + one
sum2
```

**Placeholders**

When training complicated models you need to get large amounts of data & u can't get all of it in one shot. This placeholders offers u a possibility to get data in smaller chunks. This is idea behind concept of placeholders.

For constants and variables, u initialize them to some value while placeholders u do not initialize to any value, u just specify what type of data it should hold. e.g. 
a = tf.placeholder(tf.float32). You give value of placeholder during execution time.

## Difference b/w tf.constant and tf.Variable

In TensorFlow, tf.constant and tf.Variable are two different types of objects used to represent tensors.

In TensorFlow 2.0, placeholders have been deprecated and replaced with the tf.keras.Input layer.

## What is Tensorflow?


- An open source Deep Learning library to develop deep learning applications.
- Can be used with any common programming language like Python, Java or R.
- Tf developed by one of Google's team called as Google Brain.



## Structure of Tensorflow

Consists of 2 main components-
1. Tensors
2. Flow

1. *Tensors*

Where the data is stored (like a data structure). They are multi-dimensional data arrays. They typically hold numerical values but they can also hold strings and text.

2. *Flow*

Which is a Computational Graph.


### Tensors

Data in form of tensors

### Computational Graph

- Once u have data in form of tensors, u need to create a computational graph. We do not execute a tf program like a normal java or python program.
- You write code to create a computational graph and run in it in a session. This session can run in a distributed mode which is primary reason for creating a computational graph.

The program can be executed on CPU or GPU

In Deep learning there are a lot of mathematical calculations performed like Matrix Multiplication (both row by column as well as element by element) in ANN.So GPU help to faster mathematical operations.

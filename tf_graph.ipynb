# Computational Graph - using tensorflow version 1

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

- A default graph is created when u instantiate tensorflow.
- In tensorflow every node is known as an operation irrespective of whether irrespective of whether it is actually an operation or a variable or a constant as shown below-

```python
graph = tf.get_default_graph()
graph
```

```python
graph.get_operations()
```

As of now graph is empty cz we did not create any nodes, any variables, no operations perfomed so a clean state.

name is a special parameter available with constants and variables & purpose of this name is primarily for visualization & does not hold any other value.

```python
a = tf.constant(10, name='a')
```

```python
operations = graph.get_operations()
operations
```

```python
b = tf.constant(20, name='b')
```

```python
operations = graph.get_operations()
operations
```

```python
# c = a+b

c = tf.add(a,b, name='c')
```

```python
operations = graph.get_operations()
operations
```

```python
d = tf.multiply(a,b, name='d')
```

```python
operations = graph.get_operations()
operations
```

```python
e = tf.multiply(c,d, name='e')
```

```python
operations = graph.get_operations()
operations
```

```python
sess = tf.Session()
```

```python
# Now you can execute any node in your graph
```

```python
print(sess.run(e))
```

```python
print(sess.run(d))
```
